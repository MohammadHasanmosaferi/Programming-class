```python
def sinx(x , address ):
    x = np.linspace(x[0], x[1], 100)
    y = np.sin(x)

    plt.figure(figsize=(8,6))
    plt.plot(x, y, label='sin(x)' , linewidth=2)
    plt.title("y = sin(x)") 
    plt.xlabel('x') 
    plt.ylabel('sin(x)') 
    plt.grid(True) 
    plt.legend() 
    high = f"{address}sin_high.png"
    plt.savefig(high, dpi=700)
    low = f"{address}sin_low.png"
    plt.savefig(low, dpi=100)
    plt.show()


sinx([-1,1] , r"C:Desktop\mohammad hasan\\" )
```

```python
def cot2(x, address):
    
    x = np.linspace(x[0], x[1], 1000)
    
    y = -(1 / np.tan(x / 4))**2
    
    
    y[np.isinf(y)] = np.nan

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label="y = -cot^2(x/4)", color='red', linewidth=2)
    
    plt.title("y = -cot^2(x/4)")
    plt.xlabel('x')
    plt.ylabel('-cot^2(x/4)')
    plt.grid(True)
    plt.legend()
    plt.ylim(-20, 5) 
    
    high = f"{address}-cot^2_high.png"
    plt.savefig(high, dpi=700)
    low = f"{address}-cot^2_low.png"
    plt.savefig(low, dpi=100)
    plt.show()


cot2([-1, 1], r"C:Desktop\mohammad hasan\\")
```

```python
def sinx(x, address):
    x = np.linspace(x[0], x[1], 1000)
    y = np.sin(x)
   
    plt.plot(x, y, label='sin(x)', linewidth=2)
    plt.xlabel('x')
    plt.ylabel('sin(x)')
    plt.grid(True)
    plt.legend()


def cot2(x, address):
    x = np.linspace(x[0], x[1], 1000)
    y = -(1 / np.tan(x / 4))**2
    y[np.isinf(y)] = np.nan
    plt.plot(x, y, label="y = -cot^2(x/4)", color='red', linewidth=2)
    plt.xlabel('x')
    plt.ylabel('-cot^2(x/4)')
    plt.grid(True)
    plt.legend()
    plt.ylim(-20, 5) 


def mixfunc(x,address):
    
    
    plt.figure(figsize=(10, 7)) 
    
    
    sinx(x, '') 
    cot2(x, '') 
    
    plt.title("mix func", fontsize=16) 
    plt.xlabel('x') 
    plt.ylabel('MIX') 
    
    
    plt.legend() 
    
    
    plt.tight_layout() 

    
    high = f"{address}mix_high.png"
    plt.savefig(high, dpi=700)
    low = f"{address}mix_low.png"
    plt.savefig(low, dpi=100)

  
    plt.show()



mixfunc([-1,1],r"C:Desktop\mohammad hasan\\" )
```
